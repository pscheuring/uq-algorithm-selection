```python
MODEL_MAP = {
    "mcdropout": "MC Dropout",
    "evidential": "DER",
    "bbb": r"\makecell[c]{Bayes \\ by Backprop}",
    "ensemble": "Deep Ensemble",
}

MODEL_ORDER = [
    r"\makecell[c]{Bayes \\ by Backprop}",
    "MC Dropout",
    "Deep Ensemble",
    "DER",
]

METRIC_RENAME = {
    "aleatoric_rank_corr_median": " Rank corr. aleatoric",
    "auroc_ood_median": " AUROC OOD",
    "ence_median": " ENCE",
    "infer_times_median": " Infer time (s)",
    "train_times_median": " Train time (s)",
    "mae_aleatoric_median": " MAE aleatoric",
    "mae_prediction_median": " MAE target",
    "nll_id_median": " NLL (ID)",
    "nll_ood_median": " NLL (incl. OOD)",
}

METRIC_COL_ORDER = [
    " NLL (ID)",
    " NLL (incl. OOD)",
    " ENCE",
    " MAE aleatoric",
    " Rank corr. aleatoric",
    " AUROC OOD",
    " MAE target",
    " Train time (s)",
    " Infer time (s)",
]

# Columns that should get a * (training interval metrics)
ID_INTERVAL_COLS = [
    " MAE aleatoric",
    " Rank corr. aleatoric",
    " MAE target",
    " NLL (ID)",
    " ENCE",
]

COLUMN_FORMAT_11 = (
    r">{\centering\arraybackslash}m{2.3cm}|"
    r">{\centering\arraybackslash}m{1.4cm}|"
    r">{\centering\arraybackslash}m{1.8cm}|"
    r">{\centering\arraybackslash}m{1.8cm}|"
    r">{\centering\arraybackslash}m{1.8cm}|"
    r">{\centering\arraybackslash}m{1.8cm}|"
    r">{\centering\arraybackslash}m{1.8cm}|"
    r">{\centering\arraybackslash}m{1.8cm}|"
    r">{\centering\arraybackslash}m{1.8cm}|"
    r">{\centering\arraybackslash}m{1.8cm}|"
    r">{\centering\arraybackslash}m{1.8cm}"
)



def uq_table_postprocess_and_export(
    df: pd.DataFrame,
    design_col: str,
    results_dir: Path,
    out_subdir: str,
    out_filename: str,
    auroc_ood_note: str = None,
) -> pd.DataFrame:
    df = df.copy()

    # Sort by model order and then by design column
    df["__model_order"] = pd.Categorical(
        df["UQ model"],
        categories=MODEL_ORDER,
        ordered=True,
    )
    df = df.sort_values(["__model_order", design_col]).reset_index(drop=True)
    df.drop(columns="__model_order", inplace=True)

    # Group info for UQ model (for multirow + midrule later)
    group_dict = df.groupby("UQ model", sort=False).indices
    model_names_in_order = list(group_dict.keys())

    # DataFrame row indices where a model block ends (except the last model)
    boundary_row_indices = []
    for name in model_names_in_order[:-1]:
        idxs = sorted(group_dict[name])
        boundary_row_indices.append(idxs[-1])

    # Multirow for "UQ model"
    uq_col_idx = df.columns.get_loc("UQ model")
    for name, idxs in group_dict.items():
        idxs = sorted(idxs)
        span = len(idxs)
        if span == 0:
            continue
        # First row of the group gets \multirow
        df.iloc[idxs[0], uq_col_idx] = rf"\multirow{{{span}}}{{*}}{{{name}}}"
        # All other rows in that group are empty in this column
        for i in idxs[1:]:
            df.iloc[i, uq_col_idx] = ""

    # Column order: UQ model | design_col | metrics
    final_cols = ["UQ model", design_col] + METRIC_COL_ORDER
    df = df[final_cols]

    # Cast floats for the float_format; ints stay ints (no decimals for them)
    for c in df.columns:
        if ptypes.is_float_dtype(df[c]):
            df[c] = df[c].astype(float)

    # ------------------------------------------------------------------
    # Highlight best values:
    #   - per column: global best -> bold
    #   - per model block & column: best -> underline
    #   - both -> bold + underline
    # For " ..." metrics higher is better, for " ..." lower is better.
    # ------------------------------------------------------------------
    metric_cols = [c for c in df.columns if c not in ("UQ model", design_col)]

    global_best_positions = set()
    model_best_positions = set()

    def higher_is_better(col_name: str) -> bool:
        """Return True if higher values are better for this metric."""
        # Use arrows in the pretty name: " ..." -> higher is better
        # Everything else (" ...") is lower-is-better.
        return col_name.strip().startswith("")

    # Global best per column
    for col in metric_cols:
        if not ptypes.is_numeric_dtype(df[col]):
            continue

        vals = df[col].to_numpy(dtype=float)
        if vals.size == 0:
            continue

        finite_mask = np.isfinite(vals)
        if not finite_mask.any():
            continue

        finite_vals = vals[finite_mask]

        if higher_is_better(col):
            best_val = finite_vals.max()
        else:
            best_val = finite_vals.min()

        for i, v in enumerate(vals):
            if finite_mask[i] and np.isclose(v, best_val, rtol=1e-9, atol=1e-12):
                global_best_positions.add((i, col))

    # Best per model and column
    for col in metric_cols:
        if not ptypes.is_numeric_dtype(df[col]):
            continue

        col_vals = df[col].to_numpy(dtype=float)

        for name, idxs in group_dict.items():
            idxs_sorted = sorted(idxs)
            group_vals = col_vals[idxs_sorted]
            finite = np.isfinite(group_vals)
            if not finite.any():
                continue

            finite_vals = group_vals[finite]

            if higher_is_better(col):
                best_group = finite_vals.max()
            else:
                best_group = finite_vals.min()

            for pos_in_group, row_idx in enumerate(idxs_sorted):
                if finite[pos_in_group] and np.isclose(
                    group_vals[pos_in_group],
                    best_group,
                    rtol=1e-9,
                    atol=1e-12,
                ):
                    model_best_positions.add((row_idx, col))

    def format_num(val: float) -> str:
        """Format numeric values consistently with 5 decimal places."""
        return f"{val:.5f}"

    # Apply LaTeX formatting to selected cells
    target_positions = global_best_positions.union(model_best_positions)
    for (row_idx, col) in target_positions:
        val = df.at[row_idx, col]
        try:
            v_float = float(val)
        except (TypeError, ValueError):
            # Skip non-numeric cells
            continue

        s = format_num(v_float)
        is_global = (row_idx, col) in global_best_positions
        is_model = (row_idx, col) in model_best_positions

        if is_global and is_model:
            s = rf"\textbf{{\underline{{{s}}}}}"
        elif is_global:
            s = rf"\textbf{{{s}}}"
        elif is_model:
            s = rf"\underline{{{s}}}"

        df.at[row_idx, col] = s

    # Add \tnote{*} to training-interval columns (after highlighting)
    df = df.rename(
        columns={
            col: rf"{col}\tnote{{*}}" if col in ID_INTERVAL_COLS else col
            for col in df.columns
        }
    )
    if auroc_ood_note is not None:
        df = df.rename(
            columns={
                col: (rf"{col}\tnote{{**}}" if col == " AUROC OOD" else col)
                for col in df.columns
            }
        )

    # ------------------------------------------------------------------
    # LaTeX export
    # ------------------------------------------------------------------
    out_dir = results_dir / out_subdir
    out_dir.mkdir(parents=True, exist_ok=True)
    out_path = out_dir / out_filename

    latex_str = df.to_latex(
        index=False,
        escape=False,                      # for $...$, \multirow, \makecell, etc.
        float_format=lambda x: f"{x:.5f}", # 5 decimals for plain floats
        column_format=COLUMN_FORMAT_11,
    )

    # Add \midrule between model blocks
    lines = latex_str.splitlines()

    # Indices of all table rows (header + data)
    row_line_indices = []
    for i, line in enumerate(lines):
        if "&" in line and line.strip().endswith(r"\\"):
            row_line_indices.append(i)

    # row_line_indices[0] is the header row
    for df_idx in boundary_row_indices:
        if 1 + df_idx < len(row_line_indices):
            li = row_line_indices[1 + df_idx]  # +1 to skip header
            lines[li] = lines[li].replace(
                r"\\",
                r"\\ \midrule"
            )

    # Tablenotes block (ENCE now has a star, so only NLL incl. OOD + AUROC OOD are full-test)
    notes_lines = [
        r"\begin{tablenotes}",
        r"    \item[*] Evaluation metrics marked with \textsuperscript{*} are computed on the training interval $[-4,4]$. NLL (incl.\ OOD) and AUROC OOD are computed on the full test interval $[-6,6]$."
    ]

    # Optional AUROC OOD note with ** marker
    if auroc_ood_note is not None:
        notes_lines.append(f"    \\item[**] {auroc_ood_note}")

    notes_lines.append(r"\end{tablenotes}")

    notes_block = "\n".join(notes_lines)

    # Insert notes directly after \end{tabular}
    for i, line in enumerate(lines):
        if line.strip() == r"\end{tabular}":
            lines.insert(i + 1, notes_block)
            break

    latex_str = "\n".join(lines)

    with open(out_path, "w", encoding="utf-8") as f:
        f.write(latex_str)

    print(f"Wrote LaTeX table to {out_path}")
    return df
```

## Experiment $E_1$

```python
def make_exp1_table(eval_df: pd.DataFrame, results_dir: Path) -> pd.DataFrame:
    exp_raw = eval_df[eval_df["experiment_name"] == "exp1_n_unique"].copy()

    median_cols = [c for c in eval_df.columns if c.endswith("_median")]
    df = exp_raw[["model_name", "train_instances"] + median_cols].copy()

    # rename metrics
    df.rename(columns=METRIC_RENAME, inplace=True)

    # UQ model mapping
    df["UQ model"] = df["model_name"].map(MODEL_MAP).fillna(df["model_name"])
    df.drop(columns=["model_name"], inplace=True)

    # design column: N_unique
    design_col = r"$N_{\text{unique}}$"
    df[design_col] = df["train_instances"].astype(int)
    df.drop(columns=["train_instances"], inplace=True)

    # ensure all metric columns exist
    for col in METRIC_COL_ORDER:
        if col not in df.columns:
            df[col] = float("nan")

    df = df[["UQ model", design_col] + METRIC_COL_ORDER]

    return uq_table_postprocess_and_export(
        df,
        design_col=design_col,
        results_dir=results_dir,
        out_subdir="exp1_n_unique",
        out_filename="exp1_metrics.tex",
    )

exp1_df = make_exp1_table(eval_df, RESULTS_DIR)
```

## Experiment $E_3$

```python
def make_exp3_table(eval_df: pd.DataFrame, results_dir: Path) -> pd.DataFrame:
    # base exp3 rows
    exp3_raw = eval_df[eval_df["experiment_name"] == "exp3_inbetween_ood"].copy()
    # exp1 rows with N_unique = 1000
    exp1_raw = eval_df[eval_df["experiment_name"] == "exp1_n_unique"].copy()
    exp1_1000 = exp1_raw[exp1_raw["train_instances"] == 1000].copy()

    median_cols = [c for c in eval_df.columns if c.endswith("_median")]
    base_cols = ["model_name", "train_interval"]

    exp3_filtered = exp3_raw[base_cols + median_cols].copy()
    exp1_filtered_1000 = exp1_1000[base_cols + median_cols].copy()

    df = pd.concat([exp3_filtered, exp1_filtered_1000], ignore_index=True)

    df.rename(columns=METRIC_RENAME, inplace=True)

    df["UQ model"] = df["model_name"].map(MODEL_MAP).fillna(df["model_name"])
    df.drop(columns=["model_name"], inplace=True)

    # In-between OOD column with \cmark / \xmark
    design_col = "In-between OOD"
    target_repr = str([[-4.0, -1.0], [1.0, 4.0]])

    def _ood_flag(v):
        return r"\cmark" if str(v) == target_repr else r"\xmark"

    df[design_col] = df["train_interval"].apply(_ood_flag)
    df.drop(columns=["train_interval"], inplace=True)

    for col in METRIC_COL_ORDER:
        if col not in df.columns:
            df[col] = float("nan")

    df = df[["UQ model", design_col] + METRIC_COL_ORDER]

    footnote_string = "In the in-between OOD setting, both outer OOD samples (outside the training domain) and in-between OOD samples (lying in the gap between the training intervals) are treated as OOD and are fully included in the OOD evaluation."

    return uq_table_postprocess_and_export(
        df,
        design_col=design_col,
        results_dir=results_dir,
        out_subdir="exp3_inbetween_ood",
        out_filename="exp3_metrics.tex",
        auroc_ood_note=footnote_string,
    )

exp3_df = make_exp3_table(eval_df, RESULTS_DIR)
```

## Experiment $E_4$

```python
def make_exp4_table(eval_df: pd.DataFrame, results_dir: Path) -> pd.DataFrame:
    exp_raw = eval_df[eval_df["experiment_name"] == "exp4_two_feat"].copy()

    median_cols = [c for c in eval_df.columns if c.endswith("_median")]
    df = exp_raw[["model_name", "train_instances"] + median_cols].copy()

    df.rename(columns=METRIC_RENAME, inplace=True)

    df["UQ model"] = df["model_name"].map(MODEL_MAP).fillna(df["model_name"])
    df.drop(columns=["model_name"], inplace=True)

    design_col = r"$N_{\text{unique}}$"
    df[design_col] = df["train_instances"].astype(int)
    df.drop(columns=["train_instances"], inplace=True)

    for col in METRIC_COL_ORDER:
        if col not in df.columns:
            df[col] = float("nan")

    df = df[["UQ model", design_col] + METRIC_COL_ORDER]

    footnote_string = "In the 2D setting, only samples that are unambiguously ID (both coordinates within the training interval) or unambiguously out-of-distribution (both coordinates simultaneously below or above the training interval) are included in the OOD evaluation; mixed cases are excluded."

    return uq_table_postprocess_and_export(
        df,
        design_col=design_col,
        results_dir=results_dir,
        out_subdir="exp4_two_feat",
        out_filename="exp4_metrics.tex",
        auroc_ood_note=footnote_string,
    )

exp4_df = make_exp4_table(eval_df, RESULTS_DIR)
```

## Experiment $E_5$

```python
def make_exp5_table(eval_df: pd.DataFrame, results_dir: Path) -> pd.DataFrame:
    exp5_raw = eval_df[eval_df["experiment_name"].str.startswith("exp5")].copy()
    exp1_raw = eval_df[eval_df["experiment_name"] == "exp1_n_unique"].copy()
    exp1_1000 = exp1_raw[exp1_raw["train_instances"] == 1000].copy()

    median_cols = [c for c in eval_df.columns if c.endswith("_median")]
    base_cols = ["experiment_name", "model_name", "train_instances"]

    filtered_exp5 = exp5_raw[base_cols + median_cols].copy()
    exp1_filtered_1000 = exp1_1000[base_cols + median_cols].copy()

    df = pd.concat([filtered_exp5, exp1_filtered_1000], ignore_index=True)

    df.rename(columns=METRIC_RENAME, inplace=True)

    df["UQ model"] = df["model_name"].map(MODEL_MAP).fillna(df["model_name"])
    df.drop(columns=["model_name"], inplace=True)

    gen_map = {
        "exp5_multi_target/distinct": r"$\mathcal{G}_{\text{multi-target}}^{(2)}$",
        "exp5_multi_target/similar": r"$\mathcal{G}_{\text{multi-target}}^{(1)}$",
        "exp1_n_unique": r"$\mathcal{G}_{\text{single}}$",
    }
    design_col = "Data generator"
    df[design_col] = df["experiment_name"].map(gen_map).fillna(df["experiment_name"])

    # train_instances is not shown in the table
    df.drop(columns=["experiment_name", "train_instances"], inplace=True)

    for col in METRIC_COL_ORDER:
        if col not in df.columns:
            df[col] = float("nan")

    df = df[["UQ model", design_col] + METRIC_COL_ORDER]

    return uq_table_postprocess_and_export(
        df,
        design_col=design_col,
        results_dir=results_dir,
        out_subdir="exp5_multi_target",
        out_filename="exp5_metrics.tex",
    )

exp5_df = make_exp5_table(eval_df, RESULTS_DIR)
```
